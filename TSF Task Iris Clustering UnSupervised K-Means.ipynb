Since Our Dataset has equibalanced distance on both axis then we dont need to MinMax Scaler in this

```python
from sklearn.cluster import KMeans
k_means=KMeans(n_clusters=2)
k_means
y_predict=k_means.fit_predict(data[['SepalLengthCm','SepalWidthCm']])

sepalCluster=data.iloc[:,[0,1,2]]
sepalCluster['Cluster']=y_predict

data1=sepalCluster[sepalCluster.Cluster==0]
data2=sepalCluster[sepalCluster.Cluster==1]

plt.scatter(data1.SepalLengthCm,data1.SepalWidthCm,color="Red")
plt.scatter(data2.SepalLengthCm,data2.SepalWidthCm,color="Green")

plt.xlabel('SepalLengthCm')
plt.ylabel('SepalWidthCm')
plt.legend()
```

Since Checking K_Means Clusters manually is so time taking so we go for the another technique to determine the Cluster i.e $Elbow  Technique$

```python
sse=[] #Sum of square errors
for k in range(1,10,1):
    km=KMeans(n_clusters=k)
    km.fit(data[['SepalLengthCm','SepalWidthCm']])
    sse.append(km.inertia_)
plt.xlabel("K")
plt.ylabel("Sum Of Square Errors")
plt.plot(range(1,10,1),sse)
plt.show()
sse
```
